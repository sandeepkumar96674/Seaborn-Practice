### Seaborn Tutorial

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# correlation with heatmap
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#joinplot
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='time')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

## Dist plot

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True,bins=10)
```

# Categorical Plot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('sex',data=df,hue='time')
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

## Bar plot

```python
sns.barplot(x='sex',y='size',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

## Box Plot

```python
sns.boxplot('sex','size',data=df)
```

```python
sns.boxplot('day','size',data=df)
```

```python
sns.boxplot('day','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
